````python
string = "qA21324bibbbB#B$IBbijBHBH#biwbijwbrjfbwijrbjb43i25bbjibijbBBHBHBDBI#J$@R4ih4j2ijijib"

# Alphanumeric
# Alphabetical
# Digits
# Lowercase
# Uppercase

has_alphanumeric = False
has_alphabetical = False
has_digit = False
has_lowercase = False
has_uppercase = False

for char in string:
    if char.isalnum():
        has_alphanumeric = True
    if char.isalpha():
        has_alphabetical = True
    if char.isdigit():
        has_digit = True
    if char.islower():
        has_lowercase = True
    if char.isupper():
        has_uppercase = True

print(has_alphanumeric)
print(has_alphabetical) 
print(has_digit)
print(has_lowercase)
print(has_uppercase)
````

````python
number = 17
width = len(bin(number))-2 # -2 for the 0b

for i in range(1,number+1):
        oct_ = oct(i)[2:].rjust(width)
        hex_ = hex(i).upper()[2:].rjust(width)
        bin_ = bin(i)[2:].rjust(width)
        print(str(i).rjust(width), oct_, hex_, bin_)
````

- **`rjust(width)`**: Right-justifies a string within a field of given width by padding with spaces on the left. Useful for aligning columns of text to the right.
- **`ljust(width)`**: Left-justifies a string within a field of given width by padding with spaces on the right. Useful for aligning columns of text to the left.
- Both methods return a new padded string and can take an optional fill character as a second argument (default is space).

Example:
```python
"hello".rjust(10)   # Returns "     hello"
"hello".ljust(10)   # Returns "hello     "
```

\* is the args (collecting) and unpacking operator

````python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import defaultdict

n,m = input().split()

n_dict = defaultdict(list)

for i in range(1, int(n)+1):
    n_dict[input()].append(i)
    
for _ in range(int(m)):
    result = n_dict.get(input(), [-1])
    print(*result)
````

````python
from collections import Counter

x = list("2 3 4 5 6 8 7 6 5 18".split())

y = Counter(x)

print(y)
y['5']-=1
y['18']-=2
y+=Counter()
print(y)
````

namedtuple parses the string internally! It splits 'x,y' on commas (or spaces) and extracts the field names.

So these all create the same thing:

```python
Point = namedtuple('Point', 'x,y')       # comma-separated string
Point = namedtuple('Point', 'x y')       # space-separated string
Point = namedtuple('Point', ['x', 'y'])  # list of names
```

````python
from collections import namedtuple

n = 5
columns = "MARKS CLASS NAME ID"
students = []

input_students = ["92,2,Calum,1",        
"82,5,Scott,2",
"94,2,Jason,3",        
"55,8,Glenn,4",        
"82,2,Fergus,5"]

Student = namedtuple('Student', columns)

for i in range(n):
    students.append(Student(*input_students[i].split(",")))
    # students.append()

round(sum(int(student.MARKS) for student in students)/n,2)
````
