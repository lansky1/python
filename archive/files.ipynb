### I/O with Files

````python
%%writefile myfile.txt
Hello this is a text file
this is the second line
this is the third line
````

````python
myfile = open('myfile.txt')
````

````python
myfile.read()
````

````python
myfile.read()
````

The cursor iterates to the end, hence the next read is an empty string. Need to seek it back to the start.

````python
myfile.seek(0)
````

````python
myfile.read()
````

| Method | Returns | Behavior |
|--------|---------|----------|
| **readline()** | Single string | Reads ONE line (including `\n`) |
| **readlines()** | List of strings | Reads ALL lines into a list |

````python
myfile.seek(0)
myfile.readlines()
````

````python
myfile.close()
````

for auto closure we use with

```
with open('myfile.txt') as myfile:
    contents = myfile.read()
```

---

Permissions

| Mode | File must already exist? | Old content deleted when opened? | Can read? | Can write? | Where writing happens |
| ---- | ------------------------ | -------------------------------- | --------- | ---------- | --------------------- |
| `r`  | Yes                      | No                               | Yes       | No         | Not allowed |
| `w`  | No                       | Yes                              | No        | Yes        | Starts from beginning after clearing file |
| `a`  | No                       | No                               | No        | Yes        | Always at the end |
| `r+` | Yes                      | No                               | Yes       | Yes        | At current cursor position |
| `w+` | No                       | Yes                              | Yes       | Yes        | Starts from beginning after clearing file |

````python
with open('myfile.txt', mode='w') as myfile:
    contents = myfile.read()
````

'w' immediately truncates the file to 0 bytes as soon as it’s opened.

````python
with open('myfile.txt', mode='r') as myfile:
    print(myfile.read())
````
