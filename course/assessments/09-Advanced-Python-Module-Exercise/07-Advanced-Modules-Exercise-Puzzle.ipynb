# Better approaches

os.walk(): Cleaner for recursive directory traversal<br>
pathlib: More modern and elegant for file operations

Mine: 4.44 ms ± 131 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)<br>
os.walk(): 5.23 ms ± 1.19 ms per loop (mean ± std. dev. of 7 runs, 10 loops each)<br>
pathlib: 5.55 ms ± 121 μs per loop (mean ± std. dev. of 7 runs, 100 loops each)

```python
%%timeit
import os
import re

for dirpath, dirnames, filenames in os.walk('extracted_content'):
    for filename in filenames:
        full_file_path = os.path.join(dirpath, filename)
        with open(full_file_path, 'r') as f:
            content = f.read()
            result = re.search(r"\d{3}-\d{3}-\d{4}", content)
            if result:
                print(result)
                break  # Optional: stop after finding it
```

```python
%%timeit
from pathlib import Path
import re

pattern = re.compile(r"\d{3}-\d{3}-\d{4}")

for file_path in Path('extracted_content').rglob('*'):
    if file_path.is_file():
        with open(file_path, 'r') as f:
            result = pattern.search(f.read())
            if result:
                print(result)
```
